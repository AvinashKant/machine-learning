```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
```

```python
area = np.array([800, 1000, 1200, 1500, 1800, 2000, 2200, 2500])
bedrooms = np.array([2, 2, 3, 3, 4, 4, 4, 5])

price = np.array([200000, 250000, 300000, 360000, 420000, 460000, 500000, 580000])

X = np.column_stack((area, bedrooms))

model = LinearRegression()
model.fit(X, price)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(area, bedrooms, price, marker='o')

ax.set_xlabel("Area (sq ft)")
ax.set_ylabel("Bedrooms")
ax.set_zlabel("House Price")

ax.set_title("3D House Price Prediction")

plt.show()
```
