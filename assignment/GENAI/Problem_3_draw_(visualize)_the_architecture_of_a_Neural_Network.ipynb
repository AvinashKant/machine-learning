```python
import matplotlib.pyplot as plt
```

```python
class NeuralNetworkVisualizer:

    def __init__(self, layers, input_labels, output_label):
        self.layers = layers
        self.input_labels = input_labels
        self.output_label = output_label

    def draw_network(self):

        fig, ax = plt.subplots(figsize=(18,7))

        layer_positions = []

        for i, neurons in enumerate(self.layers):

            y_positions = list(range(neurons))
            x_positions = [i]*neurons
            layer_positions.append(list(zip(x_positions, y_positions)))

            for j,(x,y) in enumerate(zip(x_positions,y_positions)):

                circle = plt.Circle((x,y),0.25,fc='lightblue',ec='black')
                ax.add_patch(circle)

                if i == 0:
                    ax.text(x-0.6, y, self.input_labels[j], fontsize=9)

                if i == len(self.layers)-1:
                    ax.text(x+0.4, y, self.output_label, fontsize=11)

        for i in range(len(layer_positions)-1):

            for (x1,y1) in layer_positions[i]:
                for (x2,y2) in layer_positions[i+1]:
                    ax.plot([x1,x2],[y1,y2])

        ax.text(0, max(self.layers)+0.5, "Input Layer", ha='center', fontsize=12)
        ax.text(1, max(self.layers)+0.5, "Hidden Layer 1", ha='center', fontsize=12)
        ax.text(2, max(self.layers)+0.5, "Hidden Layer 2", ha='center', fontsize=12)
        ax.text(3, max(self.layers)+0.5, "Output Layer", ha='center', fontsize=12)

        ax.set_title("Neural Network Architecture for Credit Card Fraud Detection", fontsize=14)

        ax.set_xlim(-1, len(self.layers))
        ax.set_ylim(-1, max(self.layers)+1)

        ax.axis('off')

        plt.show()
```

```python
input_labels = [
"TransactionAmount",
"TransactionTime",
"MerchantCategory",
"CustomerAge",
"AccountBalance",
"NumberOfTransactionsToday"
]

layers = [6,8,4,1]

output_label = "Fraud Probability"

visualizer = NeuralNetworkVisualizer(layers, input_labels, output_label)
visualizer.draw_network()
```
